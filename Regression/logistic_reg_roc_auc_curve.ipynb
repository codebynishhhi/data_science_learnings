```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
X, y = make_classification(n_samples=1000, n_classes=2, random_state=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

generate a dummy model that will give output as 0 always

```python
# dummy model will give output as 0 by deafult has been set - all output as 0
dummy_model_prob = [0 for _ in range(len(y_test))]
dummy_model_prob
```

# Create a Basic Logistic Model

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.fit(X_test, y_test)
```

```python
# test data prediction based on probability
model_prob = model.predict_proba(X_test)
```

```python
# focus on only 1 type of output - positive output
model_prob = model_prob[:,1]
```

Calculate the scores

```python
dummy_model_auc = roc_auc_score(y_test, dummy_model_prob)
model_auc = roc_auc_score(y_test, model_prob)
print(dummy_model_auc)
print(model_auc)
```

# Calculate the ROC Curve

```python
# Unpack into three variables: fpr, tpr, and thresholds
dummy_fpr, dummy_tpr, _ = roc_curve(y_test, dummy_model_prob)
model_fpr, model_tpr, thresholds = roc_curve(y_test, model_prob)
```

```python
plt.plot(dummy_fpr, dummy_tpr, label='Dummy Model ROC curve')
plt.plot(model_fpr, model_tpr, label='Model ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

# Get the threshold value for each data point in the curve from the ROC curve

```python
fig = plt.figure(figsize=(20,25))
plt.plot(dummy_fpr, dummy_tpr,linestyle='--' ,label='Dummy Model ROC curve')
plt.plot(model_fpr, model_tpr, marker='.',label='Model ROC curve')

# This line creates an axis object that covers the whole figure
ax = fig.add_subplot(111)

# Annotates points on the graph with their corresponding threshold value
for xyz in zip(model_fpr, model_tpr, thresholds):
    # xyz[0] is the FPR, xyz[1] is the TPR, xyz[2] is the threshold
    ax.annotate('%s' %np.round(xyz[2],2), xy=(xyz[0], xyz[1]))

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
